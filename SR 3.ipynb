```python
#Importing  Libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as npy
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\user\Desktop\Tp2 activity 5\Informal_Settlement_Programme_-1898186875815614491.csv')
```

```python
# Bar chart: Number of Settlements per Region
region_counts = df['REGION'].value_counts() 

plt.figure(figsize=(10, 6))
region_counts.plot(kind='bar')
plt.title('Number of Settlements per Region')
plt.xlabel('Region')
plt.ylabel('Number of Settlements')
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Pie Chart: Proportion of Settlements with Different Water Supply Types
water_supply_counts = df['WATER_SUPP'].loc[1:11].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(water_supply_counts, labels=water_supply_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Settlements with Different Water Supply Types')
plt.axis('equal')
plt.show()
```

```python
# 3. Histogram: Distribution of Estimated Household Counts
plt.figure(figsize=(10, 6))
plt.hist(df['EST_HHC'].dropna(), bins=20, edgecolor='k')
plt.title('Distribution of Estimated Household Counts')
plt.xlabel('Estimated Household Counts')
plt.ylabel('Frequency')
plt.show()
```

```python
# Advanced Charts
# Select only numerical columns for the correlation matrix
numerical_columns = df.select_dtypes(include=[npy.number])
plt.figure(figsize=(12, 8))
sns.heatmap(numerical_columns.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```

```python
# Box plot: Estimated Household Counts by Region
plt.figure(figsize=(12, 8))
sns.boxplot(x='REGION', y='EST_HHC', data=df)
plt.title('Estimated Household Counts by Region')
plt.xlabel('Region')
plt.ylabel('Estimated Household Counts')
plt.xticks(rotation=45)
plt.show()
```
