```python
#Importing  Libraries
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\user\Desktop\Tp2 activity 5\Informal_Settlement_Programme_-1898186875815614491.csv')
```

```python
# Insertion Sort Algorithm
def insertion_sort(array):
    for i in range(1, len(array)):
        key = array[i]
        j = i - 1
        # Move elements of array[0..i-1], that are greater than key, to one position ahead of their current position
        while j >= 0 and key < array[j]:
            array[j + 1] = array[j]
            j -= 1
        array[j + 1] = key
    return array

# Sort the 'EST_HHC' column
est_hhc_sorted = insertion_sort(df['EST_HHC'].dropna().tolist())

# Add sorted values back to DataFrame
df_sorted = df.copy()
df_sorted['EST_HHC'] = pd.Series(est_hhc_sorted)

# Display the first 10 rows of the sorted DataFrame
print(df_sorted.head(10))

# Displaying original and sorted EST_HHC for comparison
print("Original EST_HHC:")
print(df['EST_HHC'].head(10).tolist())
print("Sorted EST_HHC:")
print(est_hhc_sorted[:10])
```
